```python
##Task 2: Dimensionality Reduction with PCA:
```

```python
from sklearn.preprocessing import StandardScaler
numerical_cols = df.select_dtypes(include=['float64', 'int64']).columns
scaler = StandardScaler()
df[numerical_cols] = scaler.fit_transform(df[numerical_cols])
```

```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
pca = PCA()
pca.fit(df[numerical_cols])
explained_variance = pca.explained_variance_ratio_
plt.figure(figsize=(8, 5))
plt.plot(range(1, len(explained_variance) + 1), explained_variance, marker='o', linestyle='--')
plt.title('Explained Variance by Principal Components')
plt.xlabel('Principal Component Index')
plt.ylabel('Variance Explained')
plt.grid()
plt.show()
cumulative_variance = explained_variance.cumsum()
n_components = (cumulative_variance >= 0.95).argmax() + 1
print(f'Number of components to explain 95% variance: {n_components}')
pca_reduced = PCA(n_components=n_components)
df_pca = pca_reduced.fit_transform(df[numerical_cols])
df_pca = pd.DataFrame(df_pca, columns=[f'PC{i+1}' for i in range(n_components)])
print(df_pca)
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(df[numerical_cols])
pca = PCA()
pca.fit(X_scaled)
explained_variance = pca.explained_variance_ratio_
cumulative_variance = explained_variance.cumsum()
plt.figure(figsize=(10, 6))
plt.plot(range(1, len(explained_variance) + 1), explained_variance, marker='o', linestyle='--', color='b')
plt.title('Scree Plot')
plt.xlabel('Principal Component Number')
plt.ylabel('Explained Variance')
plt.xticks(range(1, len(explained_variance) + 1))
plt.grid()
plt.show()
plt.figure(figsize=(10, 6))
plt.plot(range(1, len(cumulative_variance) + 1), cumulative_variance, marker='o', linestyle='--', color='g')
plt.axhline(y=0.90, color='r', linestyle='--') 
plt.axhline(y=0.95, color='b', linestyle='--') 
plt.title('Cumulative Explained Variance Plot')
plt.xlabel('Number of Principal Components')
plt.ylabel('Cumulative Explained Variance')
plt.xticks(range(1, len(cumulative_variance) + 1))
plt.grid()
plt.show()
```

```python
##Task 3: Clustering with Original Data:
```

```python
import os
os.environ["OMP_NUM_THREADS"] = "1"  # Avoid memory leak warning on Windows with MKL
```

```python


import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
```

```python
numerical_cols = df.select_dtypes(include=['float64', 'int64']).columns
scaler = StandardScaler()
X_scaled = scaler.fit_transform(df[numerical_cols])
inertia = []
range_n_clusters = range(1, 10)
for n_clusters in range_n_clusters:
    kmeans = KMeans(n_clusters=n_clusters, random_state=42)
    kmeans.fit(X_scaled)
    inertia.append(kmeans.inertia_)

plt.figure(figsize=(8, 4))
plt.plot(range_n_clusters, inertia, marker='o')
plt.title('Elbow Method for Optimal k')
plt.xlabel('Number of Clusters')
plt.ylabel('Inertia')
plt.xticks(range_n_clusters)
plt.grid()
plt.show()

optimal_n_clusters = 3  
kmeans = KMeans(n_clusters=optimal_n_clusters, random_state=42)
cluster_labels = kmeans.fit_predict(X_scaled)

df['Cluster'] = cluster_labels

pca = PCA(n_components=2)
pca_data = pca.fit_transform(X_scaled)
pca_df = pd.DataFrame(pca_data, columns=['PC1', 'PC2'])

df_pca = pd.concat([pca_df, pd.Series(cluster_labels, name='Cluster')], axis=1)


plt.figure(figsize=(10, 6))
sns.scatterplot(data=df_pca, x='PC1', y='PC2', hue='Cluster', palette='Set2', alpha=0.7)
plt.title('K-means Clustering Results')
plt.legend(title='Cluster')
plt.show()
```

```python
# Visual Comparison of Clustering Results

# Original Data Clustering:
# K-means clustering was applied directly to the standardized numerical features of the dataset.
# The optimal number of clusters was determined using the Elbow Method, which suggested k=3.
# These cluster labels were then assigned to the original DataFrame under the column 'Cluster'.

# PCA-Transformed Clustering:
# Principal Component Analysis (PCA) was used to reduce the dataset to two principal components (PC1 and PC2),
# allowing for effective 2D visualization of the clustering structure.
# The same cluster labels obtained from the original K-means clustering were used to color the PCA scatter plot.

# Observation:
# If the PCA-based scatter plot reveals clearly separated clusters with minimal overlap,
# this indicates that the clustering performed on the original high-dimensional space is meaningful.
# Furthermore, it suggests that PCA has preserved the intrinsic cluster structure during dimensionality reduction.
```

```python
# Similarities Between Original and PCA-Based Clustering

# - Both methods use the same clustering algorithm: K-means with 3 clusters.
# - Both are based on the same standardized numerical features from the original dataset.
# - PCA retained the core structure of the data, as shown by the clear groupings in the PCA scatter plot.
```

```python
# Benefits of Applying PCA in Clustering

# - Dimensionality Reduction: PCA reduces high-dimensional data to 2 or 3 principal components, 
#   enabling intuitive visualization of clustering results.
# - Noise Reduction: By emphasizing the most informative features, PCA helps to filter out noise 
#   and improve the stability and performance of clustering algorithms.
# - Feature Simplification: PCA is particularly useful when the original data contains redundant 
#   or highly correlated features, simplifying the feature space without significant loss of information.

# Potential Drawbacks of PCA in Clustering

# - Loss of Nonlinear Relationships: PCA is a linear transformation and may not capture complex 
#   nonlinear patterns that could be important for distinguishing clusters.
# - Altered Distance Metrics: Since K-means clustering relies on Euclidean distances, the 
#   transformation applied by PCA can modify these relationships, potentially affecting cluster assignments.
# - Risk of Information Loss: While PCA improves efficiency, it may discard subtle but important 
#   data characteristics that are relevant to clustering, especially when reducing to very few dimensions.

# Summary:
# PCA can enhance clustering by improving visualization and reducing noise, particularly in datasets 
# with high dimensionality or redundancy. However, its linear nature and potential impact on distance 
# measures mean it should be applied with careful consideration of the data structure.
```

```python
# Advantages of Using PCA
# - Reduces dimensionality, making data simpler to work with.
# - Enables easier visualization in 2D or 3D plots.

# Disadvantages of Using PCA
# - Important features may be lost during dimensionality reduction.
# - It can distort distances, which may impact clustering accuracy.

# Best Practice
# - Apply PCA when dealing with high-dimensional or noisy data.
# - Use original data when interpretability and accuracy are important.
# - Compare results from both approaches using metrics like Adjusted Rand Index or Silhouette Score 
#   to make better-informed decisions.
```

```python
# When to Use PCA Before Clustering
# - If the data has many features, especially if some are redundant or highly correlated.
# - When you want to visualize clusters in 2D or 3D for easier interpretation.

# When to Cluster on Original Data
# - If it's important to understand clusters based on the original features.
# - When the dataset has only a few features, so PCA may not add much value.

# Balanced Strategy
# - Use PCA for initial exploration and visualization.
# - Compare clustering results from both methods using evaluation metrics 
#   like Adjusted Rand Index or Silhouette Score to choose the best approach.
```
